## Exercise 3

Write a function named __inside__ that receives two input parameters. The first parameter is a list of elements and the second one is an element. The function returns `True` if the element is in the list. Otherwise returns `False`. 

* You can use the `if-else` statement
* You can use the function `in` to know if an element is inside the sequence

__Remark:__

* `inside([1, 3, 6], 1)` returns `True`
* `inside([1, 3, 6], 5)` returns `False`
* `inside([1,'Hi', 6, 'Hola'] , 'Hola')` returns `True`

```python
# Sol:
def inside(list, element):
    for i in list:
        if i == element:
            return True
    return False

    
print(inside([1,3,6],1))
print(inside([1,3,6],5))
print(inside([1,'Hi',6,'Hola'],'Hola'))

```
